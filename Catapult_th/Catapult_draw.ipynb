```python
import pandas as pd
import numpy as np
```

```python
# catapult prob
def catapult_prob(ice = 1):
    cnt = np.zeros(3000,dtype=np.float64)
    for frozen in range(399, 600):
        for L in range(175, 185):
            c = ice + frozen + 146

            lo = int(np.ceil(2.5 * (L / np.float32(0.37) - ice + 1)))
            hi = int(np.ceil(2.5 * (L / np.float32(0.23) - ice + 1)))

            for k in range(lo, hi + 1):
                v_lo = L / ( (k / 2.5) + ice - 1)
                v_hi = L / ( ((k - 1) / 2.5) + ice - 1)
                v_lo = max(v_lo, 0.23)
                v_hi = min(v_hi, 0.37)
                prob = (v_hi - v_lo) / (0.37 - 0.23)
                if prob > 0:
                    cnt[k+c] += prob
    cnt /= 2010
    return cnt

prob = catapult_prob(474)
idx = np.where(prob>0)[0]
idx[0], idx[-1] 
```
