```python
def analyese_distribution(data, t_trig=2301, show=True):
    pdf = data
    cdf = np.cumsum(pdf)
    idx = np.where((cdf > 0) & (cdf < pdf.sum()-1e-5))[0]

    print(f'出手率{cdf[t_trig]*100:.5f}%，最大出手率{cdf[idx[-1]]*100:.5f}%')
    print(f'最早出手{idx[0]}, 最晚出手{idx[-1]}')
    if show:
        # plt.figure(figsize=(5,3))
        plt.plot(idx, cdf[idx])
        plt.show()
    return idx, cdf

def analyese_distribution_name(file_name, t_trig=2301, show=True):
    data = pd.read_csv(f'{file_name}.csv')['value'].to_numpy()
    return analyese_distribution(data, t_trig, show)
```

```python
analyese_distribution_name('result', 2301, show=True)
analyese_distribution_name('738', 2301, show=True)
analyese_distribution_name('747', 2301, show=True)
analyese_distribution_name('745', 2301, show=True)
print()
```
