```python
import numpy as np
import pandas as pd
from scipy.stats import binom, geom
import matplotlib.pyplot as plt
```

```python
# 蹦极
# 二项分布
ws = 7401
giga = False
p = 1000 / (ws+1000+(5000 if giga else 0))
n = 41 * 2
x = np.arange(0, n+1)

binom_pmf = binom.pmf(x, n, p)
plt.figure(figsize=(10, 4))
plt.plot(x, binom_pmf, ms=8, label='binom pmf')
plt.title(f'binom, pmf(0) = {binom_pmf[0]*100:.2f}%, mean = {binom.mean(n, p):.2f}')
plt.show()


# 几何分布
p = binom_pmf[0] # 成功概率
x = np.arange(1, 101)

geom_pmf = geom.pmf(x, p)
plt.figure(figsize=(10, 4))
plt.plot(x, geom_pmf, ms=8, label='geom pmf')
plt.title(f'geom, p = {p*100:.2f}%, mean = {geom.mean(p):.2f}')
plt.show()

```

```python
# 铁桶
# 二项分布
ws = 16401
p = 3000 / ws
n = 41 * 2 + 50 * 18
x = np.arange(0, n+1)

binom_pmf = binom.pmf(x, n, p)
plt.figure(figsize=(10, 4))
plt.plot(x, binom_pmf, ms=8, label='binom pmf')
plt.title(f'binom, 1 - cdf(199) = {(1 - binom.cdf(199, n, p))*100:.2f}%, mean = {binom.mean(n, p):.2f}')
plt.show()

# 几何分布
p = 1 - binom.cdf(199, n, p) 
x = np.arange(1, 101)

geom_pmf = geom.pmf(x, p)
plt.figure(figsize=(10, 4))
plt.plot(x, geom_pmf, ms=8, label='geom pmf')
plt.title(f'geom, p = {p*100:.2f}%, mean = {geom.mean(p):.2f}')
plt.show()

```
