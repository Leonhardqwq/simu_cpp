```python
import random
import pandas as pd
import matplotlib.pyplot as plt

N = 1e6
N = int(N)

TOTAL_ROW = 6
# 舞+红普障报丑矿跳蹦
spwan_total_weight = 1000*8

last_pick = [0]*TOTAL_ROW
second_last_pick = [0]*TOTAL_ROW

type_list = []
row_list = []

for _ in range(N):
    type_list.append(random.randint(0,spwan_total_weight-1)<1000)
    
    weight = [1]*TOTAL_ROW
    if type_list[-1]:  weight[0]=weight[-1]=0
    
    sum_weight = sum(weight)
    wp = [1.0*w/sum_weight for w in weight]
    
    p_last = [ (6*last_pick[i]*wp[i]+6*wp[i]-3 )/4 for i in range(TOTAL_ROW) ]
    p_second_last = [ (second_last_pick[i]*wp[i]+wp[i]-1)/4 for i in range(TOTAL_ROW)]
    
    smooth_weight = [wp[i]*min(100,max(0.01,p_last[i]+p_second_last[i])) for i in range(TOTAL_ROW)]
    #print(type_list[-1],smooth_weight)
    
    row = 0
    sum_smooth_weight = sum(smooth_weight)
    
    tmp = random.random()*sum_smooth_weight
    ans = 0
    for i in range(TOTAL_ROW):
        ans = ans + smooth_weight[i]
        if tmp <=ans:   
            row = i
            break
    
    row_list.append(row)
    
    for ro in range(TOTAL_ROW):
        last_pick[ro] = last_pick[ro] + 1
        second_last_pick[ro] = second_last_pick[ro] + 1
    second_last_pick[row] = last_pick[row]
    last_pick[row] = 0

norm_list = [ row_list[i] for i in range(N) if not type_list[i]]
ww_list = [ row_list[i] for i in range(N) if type_list[i]]
dance_vc = pd.Series(type_list).value_counts().sort_index()/N  
total_vc = pd.Series(row_list).value_counts().sort_index()  /N 
ww_vc = pd.Series(ww_list).value_counts().sort_index()  /(len(ww_list))
norm_vc = pd.Series(norm_list).value_counts().sort_index()  /(len(norm_list))

print(f'舞+红普障报丑矿跳蹦\n通常波未变速\n行数:{TOTAL_ROW}\n')
print(f'所有僵尸\n{total_vc}\n')
print(f'舞王僵尸\n{ww_vc}\n')
print(f'非舞王僵尸\n{norm_vc}\n')
```
