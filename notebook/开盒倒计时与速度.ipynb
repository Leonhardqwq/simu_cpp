```python
# rnd = np.random.randint(0, 300)
# early_rnd = np.random.randint(0, 20)
rnd = 299
early_rnd = 0
v = 0.668

def get_cd(rnd: int, early_rnd: int, v: float) -> int:
    cd = rnd + 450
    if early_rnd == 0:
        cd = cd // 3
    cd = 2 * int(cd / v)
    return cd
```

```python
# vl >= threshold, vr < threshold
vl, vr = 0.66, 0.68
while vr - vl > 0.000001:
    v = (vl + vr) / 2
    cd = get_cd(rnd, early_rnd, v)
    if cd >= 754:
        vl = v
    else:
        vr = v

vl, vr, get_cd(rnd, early_rnd, vl), get_cd(rnd, early_rnd, vr)
```
