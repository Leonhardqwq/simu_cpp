```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_data(fast=True, ice_t=0):
    file_name = f'../data/iced_{"fast" if fast else "slow"}_{ice_t}.xlsx'
    return pd.read_excel(file_name)

def get_simu_data(file_name='../output.csv'):
    return pd.read_csv(file_name)['value'].to_numpy(dtype=np.float64)
```

```python
x = get_simu_data('result_simu.csv')
y = get_simu_data('results_th.csv')
x = get_simu_data()
data = get_data(ice_t=1, fast=True)
y = data['Zombie_SNORKEL_ZOMBIE'].to_numpy(dtype=np.float128)

x = x.round(3)
size = min(len(x), len(y), 5000)
x, y = x[:size], y[:size]
```

```python
diff = x - y
plt.plot(np.arange(len(diff))[:800],diff[:800])
plt.show()
```

```python
diff = np.cumsum(x) - np.cumsum(y) 
plt.plot(np.arange(len(diff))[:800],diff[:800])
plt.show()
```
