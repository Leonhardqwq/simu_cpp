```python
x = pd.read_csv("result_simu.csv")['value'].to_numpy(dtype=np.float64)
y = pd.read_csv("results_th.csv")['value'].to_numpy(dtype=np.float64)

# x = x.round(3)
size = min(len(x), len(y), 5000)
x, y = x[:size], y[:size]
```

```python
diff = x - y
plt.plot(np.arange(len(diff))[500:900],diff[500:900])
plt.show()
```

```python
diff = np.cumsum(x) - np.cumsum(y) 
plt.plot(np.arange(len(diff))[500:900],diff[500:900])
plt.show()
```
