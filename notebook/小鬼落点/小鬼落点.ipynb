```python
import itertools
def get_imp_x(plant_col, plant_type):
    plant_x = plant_col * 80 - 40
    plant_def_offset = {
        '普': np.array([+30, +50]),
        '高': np.array([+30, +70]),
        '南': np.array([+20, +80]),
        '炮头': np.array([-60, +40]),
        '炮尾': np.array([+20, +120])
    }
    plant_x_def = plant_x + plant_def_offset[plant_type]
    imp_atk = np.array([+50, +70])
    imp_x = plant_x_def - imp_atk[::-1]
    return imp_x

name = {'普': '普通', '南': '南瓜', '炮': '炮尾',}
ans_dict = {
    '场地': [],
    '列数': [],
    '植物': [],
    '小鬼落点左(int)': [],
    '小鬼落点右(int)': [],
    '巨人坐标左(float)': [],
    '巨人坐标右(float)': [],
    '不伤炸点左': [],
    '不伤炸点右': [],
}
for scene, plant_type, plant_col, lower_stack_pumpkin_before, lower_stack_pumpkin_after in itertools.product(
    ['平地', '屋顶'],
    ['普', '高', '南', '炮头', '炮尾'],
    range(1, 10),
    [False, True],
    [False, True],
):
    if plant_type == '炮头':    continue
    if plant_type == '高':      continue
    if lower_stack_pumpkin_before and (plant_type != '炮尾' and plant_type != '南'):    continue
    if lower_stack_pumpkin_after and (plant_type != '南'):     continue
    
    imp_x = get_imp_x(plant_col, plant_type)
    if lower_stack_pumpkin_before: imp_x[0] += 1
    if lower_stack_pumpkin_after:  imp_x[1] -= 1
    giga_x = df[[f"{scene}巨人左(float)", f"{scene}巨人右(float)"]].iloc[imp_x[0]:imp_x[1]+1].values
    giga_x = giga_x[~np.all(giga_x == 0, axis=1)]
    if len(giga_x) == 0:   continue
    # print(giga_x)
    
    ans_dict['场地'].append(scene)
    ans_dict['列数'].append(plant_col)
    ans_dict['植物'].append(f"{' ' if not lower_stack_pumpkin_before else '*'}{name[plant_type[0]]}{' ' if not lower_stack_pumpkin_after else '*'}")
    ans_dict['小鬼落点左(int)'].append(imp_x[0])
    ans_dict['小鬼落点右(int)'].append(imp_x[1])
    ans_dict['巨人坐标左(float)'].append(giga_x[:,0].min())
    ans_dict['巨人坐标右(float)'].append(giga_x[:,1].max())
    ans_dict['不伤炸点左'].append((int(ans_dict['巨人坐标左(float)'][-1]) - 125 - 1) / 80.0)
    ans_dict['不伤炸点右'].append((int(ans_dict['巨人坐标右(float)'][-1]) + 230 + 1) / 80.0)
    if ans_dict['不伤炸点右'][-1] > 9.0:   
        ans_dict['不伤炸点右'][-1] = np.nan
    
    print(f"{ans_dict['场地'][-1]} 投至{ans_dict['列数'][-1]}列 {ans_dict['植物'][-1]}", end=' ')
    print(f"小鬼落点: [{ans_dict['小鬼落点左(int)'][-1]:3d}, {ans_dict['小鬼落点右(int)'][-1]:3d}]", end=' ')
    print(f"巨人坐标: [{ans_dict['巨人坐标左(float)'][-1]:.10f}, {ans_dict['巨人坐标右(float)'][-1]:.10f}]", end=' ')
    print(f"左侧不伤炸点: {ans_dict['不伤炸点左'][-1]:.4f} 右侧不伤炸点: {ans_dict['不伤炸点右'][-1]:.4f}")
    # print(f"                                    贪心巨人坐标：[{giga_x[0][0]:12.10f}, {giga_x[-1][1]:12.10f}]")
ans_df = pd.DataFrame(ans_dict)
ans_df.to_excel("结论.xlsx", index=False, float_format="%.10f")
```
