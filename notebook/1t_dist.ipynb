```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv("output.csv")['value'].to_numpy()
value_count = pd.Series(x).value_counts()
```

```python
max_cd = x.max()
total_portion = 15*(max_cd) - 14*(14+1)//2
value_count_scale = value_count / len(x) * total_portion
```

```python
for i in value_count_scale.index:
    print(f'{value_count_scale[i]}')
```

```python
plt.hist(x, bins=500)  # x 是你的 numpy 数组，bins 控制分箱数
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution Histogram')
plt.show()
```
