```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']  # 任选其一存在即可
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df_fast = pd.read_excel("data/杂/万能表.xlsm", sheet_name="原速极快", usecols="A:W")
df_slow = pd.read_excel("data/杂/万能表.xlsm", sheet_name="原速极慢", usecols="A:W")
df_fast_new = pd.read_csv("data/origin/0_fast.csv")
df_slow_new = pd.read_csv("data/origin/0_slow.csv")
df_fast_new.columns = df_fast.columns
df_slow_new.columns = df_slow.columns
```

```python
def df_diff(df, df_new, draw=True):
    length = min(len(df), len(df_new))
    df = df.iloc[:length]
    df_new = df_new.iloc[:length].round(3)
    diff = df_new - df
    diff['时刻'] = df['时刻']
    for col in df.columns[1:]:
        arr_diff = diff[col].values
        arr_diff = arr_diff[~np.isnan(arr_diff)]
        arr = df[col].values[:len(arr_diff)]
        arr_new = df_new[col].values[:len(arr_diff)]
        if draw:
            plt.figure(figsize=(10, 5))
            plt.plot(arr, label="old")
            plt.plot(arr_new, label="new")
            plt.title(f"{col}")
            plt.legend()
            plt.show()
            
            plt.figure(figsize=(10, 5))
            plt.plot(range(len(arr_diff)), arr_diff, label="diff")
            plt.title(f"{col} diff")
            plt.legend()
            plt.show()
    return diff

diff_fast = df_diff(df_fast, df_fast_new, draw=True)
diff_slow = df_diff(df_slow, df_slow_new, draw=True)
```

```python
diff_fast.to_csv("data/杂/diff_fast.csv", index=False, float_format="%.3f")
diff_slow.to_csv("data/杂/diff_slow.csv", index=False, float_format="%.3f")
```

```python
folder = Path("data/origin")
for csv_file in folder.glob("*.csv"):   # 仅当前目录
    print(csv_file)
    df = pd.read_csv(csv_file)
    df.columns = df_fast.columns
    df.to_excel(f'data/万能表减速波/{csv_file.stem}.xlsx', index=False)
```
