```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']  # 任选其一存在即可
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df_fast = pd.read_excel("万能表.xlsm", sheet_name="原速极快", usecols="A:W")
df_slow = pd.read_excel("万能表.xlsm", sheet_name="原速极慢", usecols="A:W")
df_fast_new = pd.read_csv("../data/0_fast.csv")
df_slow_new = pd.read_csv("../data/0_slow.csv")
df_fast_new.columns = df_fast.columns
df_slow_new.columns = df_slow.columns
```

```python
def df_diff(df, df_new, draw=True):
    length = min(len(df), len(df_new))
    df = df.iloc[:length]
    df_new = df_new.iloc[:length].round(3)
    diff = df_new - df
    for col in df.columns[1:]:
        arr_diff = diff[col].values
        arr_diff = arr_diff[~np.isnan(arr_diff)]
        arr = df[col].values[:len(arr_diff)]
        arr_new = df_new[col].values[:len(arr_diff)]
        if draw:
            plt.figure(figsize=(10, 5))
            plt.plot(arr, label="old")
            plt.plot(arr_new, label="new")
            plt.title(f"{col}")
            plt.legend()
            plt.show()
            
            plt.figure(figsize=(10, 5))
            plt.plot(range(len(arr_diff)), arr_diff, label="diff")
            plt.title(f"{col} diff")
            plt.legend()
            plt.show()
    return diff

diff_fast = df_diff(df_fast, df_fast_new, draw=False)
diff_slow = df_diff(df_slow, df_slow_new, draw=False)
```

```python
diff_fast.to_excel("diff_fast.xlsx", index=False)
diff_slow.to_excel("diff_slow.xlsx", index=False)
```
