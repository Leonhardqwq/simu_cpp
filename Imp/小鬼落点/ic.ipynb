```python
import numpy as np
import pandas as pd
```

```python
def transform(pathname):
    df = pd.read_csv(pathname)
    df = df.rename(columns={'index': 't', 'value0': 'min', 'value1': 'max'})
    df = df[df['min']!=0]
    x_range = df['t'].iloc[0], df['t'].iloc[-1]
    ans_str = f'{x_range[0]}, {x_range[1]}\n[\n'
    for i in range(0, len(df)):
        ans_str += f'    ({df["min"].iloc[i]:.10f}, {df["max"].iloc[i]:.10f}),\n'
    ans_str += ']'
    return ans_str
```

```python
de_str = transform('de.csv')
print(de_str)
```

```python
re_str = transform('re.csv')
print(re_str)
```
